```python
# Create a list of 5 integers
numbers = [10, 25, 15, 30, 20]
print("Original list:", numbers)

# Append
numbers.append(35)
print("After append(35):", numbers)

# Extend
numbers.extend([40, 45])
print("After extend([40, 45]):", numbers)

# Insert
numbers.insert(2, 99)
print("After insert(2, 99):", numbers)

# Remove
numbers.remove(15)
print("After remove(15):", numbers)

# Pop
popped_element = numbers.pop()
print("After pop():", numbers, "| Popped element:", popped_element)

# Index
index_of_30 = numbers.index(30)
print("Index of 30:", index_of_30)

# Count
count_of_25 = numbers.count(25)
print("Count of 25:", count_of_25)

# Sort
numbers.sort()
print("After sort():", numbers)

# Reverse
numbers.reverse()
print("After reverse():", numbers)

# Clear
numbers.clear()
print("After clear():", numbers)
```

```python
# Create a tuple with 3 student names
students = ("Alice", "Bob", "Charlie")
print("Original tuple:", students)

# Try to change the second name (this will cause an error)
try:
    students[1] = "David"
except TypeError as e:
    print("Error:", e)
    print("Explanation: Tuples are immutable in Python, meaning their elements cannot be changed after creation.")
```

```python
# Create a set with duplicate values
numbers_set = {1, 2, 3, 2, 4, 1, 5, 3}
print("Set with duplicates:", numbers_set)
print("Explanation: Sets automatically remove duplicate values and store only unique elements.")
```

```python
# Create a dictionary
person = {'name': 'John', 'age': 25, 'city': 'New York'}
print("Original dictionary:", person)

# Update age
person['age'] = 26

# Add new key 'email'
person['email'] = 'john@example.com'

print("Final dictionary:", person)
```

```python
# Voting eligibility check
age = 20  # You can change this value to test different ages

if age >= 18:
    print(f"Age {age}: Eligible to vote!")
else:
    print(f"Age {age}: Not eligible to vote. Minimum voting age is 18.")
```

```python
# Grade calculator
marks = 85  # You can change this value to test different marks

if marks >= 90:
    grade = 'A'
elif marks >= 75:
    grade = 'B'
elif marks >= 50:
    grade = 'C'
else:
    grade = 'Fail'

print(f"Marks: {marks} | Grade: {grade}")
```

```python
# Number checker
number = 8  # You can change this value to test different numbers

if number > 0:
    print(f"{number} is positive")
    if number % 2 == 0:
        print(f"{number} is also even")
    else:
        print(f"{number} is odd")
elif number == 0:
    print(f"{number} is zero")
else:
    print(f"{number} is negative")
```
