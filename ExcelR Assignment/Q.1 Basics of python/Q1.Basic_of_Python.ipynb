Excerise 1

```python
num = int(input("Enter the number"))
if num > 1:

    for i in range(2, (num//2)+1):
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

Excercise 4

```python
ROOM_CHARGE_PER_DAY = 2000
DOCTOR_FEE = 1500
LAB_TEST_COST = 300
DISCOUNT_THRESHOLD = 10000
DISCOUNT_RATE = 0.10 

days_stayed = 4
lab_tests_taken = 3
medicine_charges = 2400

room_charges = days_stayed * ROOM_CHARGE_PER_DAY
lab_charges = lab_tests_taken * LAB_TEST_COST

total_before_discount = room_charges + DOCTOR_FEE + lab_charges + medicine_charges

# Apply discount if applicable
if total_before_discount > DISCOUNT_THRESHOLD:
    discount_amount = total_before_discount * DISCOUNT_RATE
    final_bill = total_before_discount - discount_amount
else:
    discount_amount = 0
    final_bill = total_before_discount


print(f"Room charges ({days_stayed} days × ₹{ROOM_CHARGE_PER_DAY}): ₹{room_charges}")
print(f"Doctor consultation fee: ₹{DOCTOR_FEE}")
print(f"Lab test charges ({lab_tests_taken} tests × ₹{LAB_TEST_COST}): ₹{lab_charges}")
print(f"Medicine charges: ₹{medicine_charges}")
print(f"Total before discount: ₹{total_before_discount}")

if discount_amount > 0:
    print(f"Discount applied ({DISCOUNT_RATE*100}%): -₹{discount_amount:.2f}")

print(f"Final bill amount: ₹{final_bill:.2f}")
```
